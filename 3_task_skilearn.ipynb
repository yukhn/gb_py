Задание 1

Импортируйте библиотеки pandas и numpy.
Загрузите "Boston House Prices dataset" из встроенных наборов данных библиотеки sklearn. Создайте датафреймы X и y из этих данных.
Разбейте эти датафреймы на тренировочные (X_train, y_train) и тестовые (X_test, y_test) с помощью функции train_test_split так, чтобы размер тестовой выборки
составлял 30% от всех данных, при этом аргумент random_state должен быть равен 42.

Создайте модель линейной регрессии под названием lr с помощью класса LinearRegression из модуля sklearn.linear_model.
Обучите модель на тренировочных данных (используйте все признаки) и сделайте предсказание на тестовых.

Вычислите R2 полученных предказаний с помощью r2_score из модуля sklearn.metrics.


```python
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston

boston = load_boston()
boston.keys()
```

```python
for i in [boston['data'], boston['target'], boston['feature_names']]:
  print(i[:2])

print(boston['DESCR'])
```

```python
X = pd.DataFrame(data=boston['data'], columns=boston['feature_names'])
y = pd.DataFrame(data=boston['target'], columns=['price'])
print(X.head(3), y.head(3))
```

```python
X.info()
print("_____________")
y.info()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
y_pred.shape
```

```python
check_test = pd.DataFrame({
    'y_test': y_test['price'],
    'y_pred': y_pred.flatten(),
})
check_test['error'] = check_test['y_pred'] - check_test['y_test']
check_test.head(10)
```

```python
from sklearn.metrics import r2_score

R2 = r2_score(y_pred, y_test)
print(R2)
```

```python
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.model_selection import train_test_split, GridSearchCV as gscv

df = pd.read_csv('creditcard.csv.zip')
df.head()
```

```python
X = df.drop('Class', 1)
y = pd.Series(df['Class'])
print(X, y)
```

```python
model = gscv(estimator=rfc(random_state=100),
    param_grid=parameters,
    scoring='roc_auc',
    cv=3)
model.fit(X_train, y_train)
```

```python
model.best_params_
```

```python
model = rfc(max_depth=6, max_features=3, n_estimators=15)
model.fit(X_train, y_train)
```

```python
from sklearn.metrics import roc_auc_score
y_pred = model.predict_proba(X_test)
y_pred_proba = y_pred[:, 1]
roc_auc_score(y_test, y_pred_proba)
```

```python
X = X.drop('target', 1)
for feature_name in high_corr:
    X[f'{feature_name}_2'] = X.apply( lambda x: pow(x[feature_name], 2), 1 )
```
